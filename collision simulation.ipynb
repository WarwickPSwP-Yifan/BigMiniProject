```python
def collisions(a, b, c): # This is for get as many collisions in the box as possible at one stage
    # x: number of collisions between A and B
    # y: number of collisions between B and C
    # z: number of collisions between A and C
    max_x=0
    max_y=0
    max_z=0
    max_collision=0 #PlaceHolders
    for x in range (a+1): # x is smaller than number of particles of A.
        if x>b:#x should be less or equal to b at most.
            break
        else:
            for y in range (b+1): # similarly, y should be less than b
                if y >c: #also,  y is less or equal to c
                    break
                else:
                    if x+y>b:# the inequality constraint.
                        break
                    else:
                                        # for z, consider x+z<=a , y+z<=c as the 2 inequalities that z takes part in, I will have a specific limitation
                        z=min(a-x,c-y) #since I want as many collisions as possible, straightly take the largest.
                        total_collision=x+y+z# To make collisions happen as many as possible, this can be done by take the total number
                        if total_collision>max_collision:
                            max_collision=total_collision
                            max_x=x
                            max_y=y
                            max_z=z
    return max_x,max_y,max_z,max_collision 

def process(a,b,c,steps):
    
    stage=[(a,b,c)] #as the initial stage
    for step in range (steps):
        x, y, z, collision=collisions(a,b,c)
        if collision==0: #No collision happens
            break
            # according to the formula in the attack phase
        a_new=a-(x+z)+2*y
        b_new=b-(x+y)+2*z
        c_new=c-(y+z)+2*x
        a=a_new
        b=b_new 
        c=c_new
        stage.append((a,b,c))
    return a,b,c,stage

def loop_or_stable(steps,stage):
    for i in range(steps):
        for j in range(i + 1, steps):
            if i != j:
                if stage[i]==stage[j]:
                    print('It will possibily be a loop')
                    return
                
                else:
                    stable=[tuples for tuples in stage if tuples.count(0)==2]
                    if len(stable)>0:
                        print('It will be stable')
                        return
                    
            
a=int(input('Please input the number of particles A'))
b=int(input('Please input the number of particles B'))
c=int(input('Please input the number of particles C'))
steps=int(input('Please input the number of steps you want to take'))

final_a,final_b,final_c,stage=process(a,b,c,steps)
for i, (a_i,b_i,c_i) in enumerate (stage):
    print(f" stage{i}:(n(A),n(B),n(C))=({a_i},{b_i},{c_i})")
loop_or_stable(steps,stage)
              
    
        
        
    


    
                    
            
        
    
```
